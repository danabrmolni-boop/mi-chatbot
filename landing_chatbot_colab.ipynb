```python
from IPython.display import HTML, display

html = r'''<!DOCTYPE html>
<html lang="es" class="scroll-smooth">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>Asistente IA — Respuestas al instante</title>
  <script src="https://cdn.tailwindcss.com"></script>
  <script>
    tailwind.config = {
      theme: {
        extend: {
          fontFamily: { sans: ['Inter', 'system-ui', 'sans-serif'] },
          borderRadius: { DEFAULT: '8px', lg: '12px' },
          boxShadow: {
            'card': '0 1px 3px 0 rgb(0 0 0 / 0.1), 0 1px 2px -1px rgb(0 0 0 / 0.1)',
            'card-hover': '0 4px 6px -1px rgb(0 0 0 / 0.1), 0 2px 4px -2px rgb(0 0 0 / 0.1)',
          }
        }
      }
    }
  </script>
  <link rel="preconnect" href="https://fonts.googleapis.com" />
  <link rel="preconnect" href="https://fonts.gstatic.com" crossorigin />
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@400;500;600;700&display=swap" rel="stylesheet" />
</head>
<body class="bg-[#0a0a0b] text-zinc-100 font-sans antialiased min-h-screen">
  <header class="sticky top-0 z-50 border-b border-white/5 bg-[#0a0a0b]/80 backdrop-blur-xl">
    <div class="max-w-6xl mx-auto px-4 sm:px-6 flex items-center justify-between h-14 sm:h-16">
      <a href="#" class="flex items-center gap-2 text-zinc-100 hover:text-white transition">
        <div class="w-8 h-8 rounded-lg bg-gradient-to-br from-violet-500 to-fuchsia-500 flex items-center justify-center">
          <svg class="w-4 h-4 text-white" fill="none" stroke="currentColor" viewBox="0 0 24 24">
            <path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M8 12h.01M12 12h.01M16 12h.01M21 12c0 4.418-4.03 8-9 8a9.863 9.863 0 01-4.255-.949L3 20l1.395-3.72C3.512 15.042 3 13.574 3 12c0-4.418 4.03-8 9-8s9 3.582 9 8z" />
          </svg>
        </div>
        <span class="font-semibold text-sm sm:text-base">Asistente IA</span>
      </a>
      <nav class="flex items-center gap-1 sm:gap-2">
        <a href="#chat" class="text-zinc-400 hover:text-zinc-100 text-sm px-3 py-2 rounded-lg hover:bg-white/5 transition">Producto</a>
        <a href="#" class="text-zinc-400 hover:text-zinc-100 text-sm px-3 py-2 rounded-lg hover:bg-white/5 transition hidden sm:inline">Recursos</a>
        <a href="#chat" class="ml-1 sm:ml-2 text-sm font-medium px-4 py-2 rounded-lg bg-white text-zinc-900 hover:bg-zinc-200 transition">Probar gratis</a>
      </nav>
    </div>
  </header>

  <main>
    <section class="pt-10 sm:pt-16 pb-12 sm:pb-16 px-4 sm:px-6">
      <div class="max-w-3xl mx-auto text-center">
        <p class="text-violet-400 text-sm font-medium mb-4">Asistencia inteligente 24/7</p>
        <h1 class="text-3xl sm:text-4xl md:text-5xl font-bold tracking-tight text-zinc-100 leading-tight">
          Respuestas al instante.<br class="hidden sm:block" />
          <span class="text-zinc-400">Sin esperas.</span>
        </h1>
        <p class="mt-4 sm:mt-6 text-zinc-400 text-base sm:text-lg max-w-xl mx-auto">
          Conecta con un asistente que entiende contexto, resuelve dudas y escala tu atención al cliente.
        </p>

        <div class="mt-8 sm:mt-10 flex flex-wrap items-center justify-center gap-x-8 gap-y-4 text-zinc-500 text-sm">
          <span class="flex items-center gap-2">
            <svg class="w-4 h-4 text-emerald-500" fill="currentColor" viewBox="0 0 20 20"><path fill-rule="evenodd" d="M10 18a8 8 0 100-16 8 8 0 000 16zm3.707-9.293a1 1 0 00-1.414-1.414L9 10.586 7.707 9.293a1 1 0 00-1.414 1.414l2 2a1 1 0 001.414 0l4-4z" clip-rule="evenodd" /></svg>
            Respuesta en segundos
          </span>
          <span class="flex items-center gap-2">
            <svg class="w-4 h-4 text-violet-500" fill="currentColor" viewBox="0 0 20 20"><path fill-rule="evenodd" d="M2.166 4.999A11.954 11.954 0 0010 1.944 11.954 11.954 0 0017.834 5c.892.65 1.814 1.51 2.166 2.5a11.954 11.954 0 01-2.166 2.5A11.954 11.954 0 0110 16.056a11.954 11.954 0 01-7.834-2.556C1.814 12.49.892 11.63 0 10.999A11.954 11.954 0 0110 1.944z" clip-rule="evenodd" /></svg>
            Privacidad garantizada
          </span>
          <span class="flex items-center gap-2">
            <svg class="w-4 h-4 text-amber-500" fill="currentColor" viewBox="0 0 20 20"><path d="M9.049 2.927c.3-.921 1.603-.921 1.902 0l1.07 3.292a1 1 0 00.95.69h3.462c.969 0 1.371 1.24.588 1.81l-2.8 2.034a1 1 0 00-.364 1.118l1.07 3.292c.3.921-.755 1.688-1.54 1.118l-2.8-2.034a1 1 0 00-1.175 0l-2.8 2.034c-.784.57-1.838-.197-1.539-1.118l1.07-3.292a1 1 0 00-.364-1.118L2.98 8.72c-.783-.57-.38-1.81.588-1.81h3.461a1 1 0 00.951-.69l1.07-3.292z" /></svg>
            Sin tarjeta de crédito
          </span>
        </div>
      </div>
    </section>

    <section id="chat" class="px-4 sm:px-6 pb-16 sm:pb-24">
      <div class="max-w-2xl mx-auto">
        <div class="rounded-xl sm:rounded-2xl bg-zinc-900/80 border border-white/5 shadow-card hover:shadow-card-hover transition-shadow overflow-hidden">
          <div class="flex items-center gap-3 px-4 sm:px-5 py-3 sm:py-4 border-b border-white/5 bg-zinc-800/50">
            <div class="w-9 h-9 rounded-lg bg-gradient-to-br from-violet-500 to-fuchsia-500 flex items-center justify-center flex-shrink-0">
              <svg class="w-4 h-4 text-white" fill="none" stroke="currentColor" viewBox="0 0 24 24">
                <path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M8 12h.01M12 12h.01M16 12h.01M21 12c0 4.418-4.03 8-9 8a9.863 9.863 0 01-4.255-.949L3 20l1.395-3.72C3.512 15.042 3 13.574 3 12c0-4.418 4.03-8 9-8s9 3.582 9 8z" />
              </svg>
            </div>
            <div class="min-w-0">
              <h2 class="font-semibold text-zinc-100 text-sm sm:text-base">Asistente</h2>
              <p class="text-xs text-zinc-500">En línea · Responde al instante</p>
            </div>
          </div>

          <div id="chat-container" class="overflow-y-auto overflow-x-hidden flex flex-col gap-4 p-4 sm:p-5 min-h-[240px] max-h-[50vh] sm:max-h-[380px] scroll-smooth" style="scrollbar-gutter: stable;">
            <div id="welcome" class="flex flex-col items-center justify-center py-8 sm:py-12 text-center">
              <div class="w-12 h-12 rounded-xl bg-zinc-800 flex items-center justify-center mb-4">
                <svg class="w-6 h-6 text-violet-400" fill="none" stroke="currentColor" viewBox="0 0 24 24">
                  <path stroke-linecap="round" stroke-linejoin="round" stroke-width="1.5" d="M8 12h.01M12 12h.01M16 12h.01M21 12c0 4.418-4.03 8-9 8a9.863 9.863 0 01-4.255-.949L3 20l1.395-3.72C3.512 15.042 3 13.574 3 12c0-4.418 4.03-8 9-8s9 3.582 9 8z" />
                </svg>
              </div>
              <p class="text-sm font-medium text-zinc-300">¿En qué puedo ayudarte?</p>
              <p class="text-xs text-zinc-500 mt-1 max-w-[220px]">Escribe abajo y pulsa Enter o el botón de enviar.</p>
            </div>
          </div>

          <div class="p-4 sm:p-5 pt-0 border-t border-white/5 bg-zinc-800/30">
            <form id="chat-form" class="flex gap-2 items-center rounded-lg border border-white/10 bg-zinc-800/80 focus-within:border-violet-500/50 focus-within:ring-1 focus-within:ring-violet-500/30 transition">
              <span class="pl-4 text-zinc-500 flex-shrink-0">
                <svg class="w-5 h-5" fill="none" stroke="currentColor" viewBox="0 0 24 24">
                  <path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M21 21l-6-6m2-5a7 7 0 11-14 0 7 7 0 0114 0z" />
                </svg>
              </span>
              <input type="text" id="message-input" placeholder="Escribe tu mensaje..." autocomplete="off" class="flex-1 min-w-0 py-3 pr-2 bg-transparent text-zinc-100 placeholder-zinc-500 text-sm sm:text-base focus:outline-none" />
              <button type="submit" id="send-btn" class="flex-shrink-0 m-1.5 w-9 h-9 rounded-lg bg-gradient-to-br from-violet-500 to-violet-600 hover:from-violet-600 hover:to-violet-700 text-white flex items-center justify-center transition disabled:opacity-50 disabled:pointer-events-none" aria-label="Enviar">
                <svg class="w-4 h-4" fill="none" stroke="currentColor" viewBox="0 0 24 24">
                  <path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M12 19l9 2-9-18-9 18 9-2zm0 0v-8" />
                </svg>
              </button>
            </form>
          </div>
        </div>
      </div>
    </section>
  </main>

  <script>
    (function () {
      var chatContainer = document.getElementById('chat-container');
      var welcome = document.getElementById('welcome');
      var form = document.getElementById('chat-form');
      var input = document.getElementById('message-input');
      var sendBtn = document.getElementById('send-btn');

      function scrollToBottom() {
        chatContainer.scrollTop = chatContainer.scrollHeight;
      }

      function addMessage(text, isUser) {
        if (welcome && welcome.parentNode) welcome.parentNode.removeChild(welcome);
        var row = document.createElement('div');
        row.className = 'flex ' + (isUser ? 'justify-end' : 'justify-start');
        var bubble = document.createElement('div');
        bubble.className = 'max-w-[85%] sm:max-w-[80%] px-4 py-2.5 rounded-lg text-sm ' +
          (isUser
            ? 'bg-gradient-to-br from-violet-500 to-violet-600 text-white'
            : 'bg-zinc-800 text-zinc-200 border border-white/5');
        bubble.textContent = text;
        row.appendChild(bubble);
        chatContainer.appendChild(row);
        scrollToBottom();
      }

      function setLoading(loading) {
        sendBtn.disabled = loading;
        input.disabled = loading;
      }

      form.addEventListener('submit', function (e) {
        e.preventDefault();
        var text = (input.value || '').trim();
        if (!text) return;
        input.value = '';
        addMessage(text, true);
        setLoading(true);
        setTimeout(function () {
          addMessage('Gracias por tu mensaje. Aquí iría la respuesta del bot.', false);
          setLoading(false);
          scrollToBottom();
        }, 600);
      });

      if (input) input.focus();
    })();
  </script>
</body>
</html>'''

# Render en la salida de Colab
# Nota: Tailwind vía CDN puede tardar 1-2s en cargar en el iframe.
display(HTML(html))

# Guarda el archivo por si lo quieres descargar / desplegar
with open('index.html', 'w', encoding='utf-8') as f:
  f.write(html)

print('Listo: renderizado arriba y guardado como index.html')
```
